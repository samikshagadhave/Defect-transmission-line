accuracy is 99%, we can apply filter image processing algorithm.

```python
from PIL import ImageFilter
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
# Load the image
image = Image.open('/content/drive/MyDrive/New_folder/DJI_0004-8.jpg')

# Resize the image
img_resize = image.resize((640, 640))

# Apply Median Filter to the image
kernel=np.array([[0,0,0],
                 [0,1,0],
                 [0,0,0]])
image_kernel=img_resize.filter(ImageFilter.Kernel(size=(3,3),kernel=kernel.flatten()))
#img = img_to_array(image_kernel)

# Save the filtered image
filtered_image_path = '/content/drive/MyDrive/New_folder/image_kernel.jpg'
image_kernel.save(filtered_image_path)

# Display the filtered image (optional)
plt.imshow(image_kernel)
plt.show()

# Run YOLOv5 detection
!python detect.py --weights '/content/drive/MyDrive/best.pt' --conf 0.06 --source {filtered_image_path} --device cpu
```

```python
from PIL import ImageFilter
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
# Load the image
image = Image.open('/content/drive/MyDrive/New_folder/DJI_0004-8.jpg')
plt.imshow(image)
plt.show()
# Resize the image
img_resize = image.resize((640, 640))

# Apply Median Filter to the image
edge_detector=np.array([[1,0,-1],
                      [0,0,0],
                      [-1,0,1]])
image_kernel=img_resize.filter(ImageFilter.Kernel(size=(3,3),kernel=edge_detector.flatten(),scale=0.2))
#img = img_to_array(image_kernel)

# Save the filtered image
filtered_image_path = '/content/drive/MyDrive/New_folder/image_kernel.jpg'
image_kernel.save(filtered_image_path)

# Display the filtered image (optional)
#plt.imshow(image_kernel)
#plt.show()

# Run YOLOv5 detection
!python detect.py --weights '/content/drive/MyDrive/best.pt' --conf 0.2  --source {filtered_image_path} --device cpu
```
